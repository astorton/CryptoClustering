```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
market_plot = df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
market_plot
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_market_data = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
scaled_df = pd.DataFrame(scaled_market_data, columns=df_market_data.columns)
scaled_df

# Copy the crypto names from the original data
scaled_df['coin_id'] = df_market_data.index

# Set the coinid column as index
scaled_df = scaled_df.set_index('coin_id')

# Display sample data
scaled_df
```

---

### Find the Best Value for k Using the Original Scaled DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia_list = []
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(scaled_df)
    inertia_list.append(k_model.inertia_)

#verify list is populated
inertia_list
```

```python
# Create a dictionary with the data to plot the Elbow curve
coin_dict = {"K":k,"Inertia":inertia_list}

# Create a DataFrame with the data to plot the Elbow curve
coin_df = pd.DataFrame(coin_dict)

#verify coin_df is created
coin_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
coin_plot = coin_df.hvplot.line(
    width=800,
    height=400,
    rot=90,
    x="K",
    y="Inertia",
    title="Coin Curve",
    xticks = k
)
coin_plot
```

#### Answer the following question: 
​
**Question:** What is the best value for `k`?
​#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** K = 4 (79.02)!

---

```python
# Initialize the K-Means model using the best value for k
k_model = KMeans(
            n_clusters=4
            )
```

```python
# Fit the K-Means model using the scaled data
k_model.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_predictions = k_model.predict(scaled_df)


# Print the resulting array of cluster values.
k_predictions
```

```python
# Create a copy of the DataFrame
scaled_df_2 = scaled_df.copy()
scaled_df_2
```

```python
# Add a new column to the DataFrame with the predicted clusters
scaled_df_2['predictions'] = k_predictions

#scaled_df_2.append(k_predictions,columns=['predictions'])

# Display sample data
scaled_df_2.head(5)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scaled_scatter = scaled_df_2.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="predictions",
    value_label = "predictions",
    hover_cols = 'coin_id'
).opts(yformatter="%.0f")

scaled_scatter
```

---

```python
# Create a PCA model instance and set `n_components=3`.
coin_pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
coin_pca_fitted = coin_pca.fit_transform(scaled_df)


# View the first five rows of the DataFrame. 
coin_pca_fitted[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
coin_pca.explained_variance_ratio_
```

```python
# Create a new DataFrame with the PCA data.

# Creating a DataFrame with the PCA data
coin_pca_df = pd.DataFrame(
    coin_pca_fitted,
    columns=["PC1", "PC2","PC3"]
    )

# Copy the crypto names from the original data
coin_pca_df['coin_id'] = df_market_data.index

# Set the coinid column as index
coin_pca_df = coin_pca_df.set_index('coin_id')


# Display sample data
coin_pca_df.head()
```

---

```python
# Create a list with the number of k-values from 1 to 11
k_2 = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia_list_2 = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k_2:
    k_model_2 = KMeans(n_clusters=i, random_state=1)
    k_model_2.fit(coin_pca_df)
    inertia_list_2.append(k_model_2.inertia_)

#verify list is populated
inertia_list_2
```

```python
# Create a dictionary with the data to plot the Elbow curve
coin_dict_2 = {"K_2":k_2,"Inertia_2":inertia_list_2}

# Create a DataFrame with the data to plot the Elbow curve
coin_df_2 = pd.DataFrame(coin_dict_2)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
coin_plot_2 = coin_df_2.hvplot.line(
    width=800,
    height=400,
    rot=90,
    x="K_2",
    y="Inertia_2",
    title="Coin Curve_2",
    xticks = k_2
)
coin_plot_2
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** K = 4 (49.665)!


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Both indicate K 4 being the best value but the actual inertia value does vary in this case 

```python
# Initialize the K-Means model using the best value for k
k_model_2 = KMeans(
            n_clusters=4
            )
```

```python
# Fit the K-Means model using the PCA data
k_model_2.fit(coin_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_predictions_2 = k_model_2.predict(coin_pca_df)

# Print the resulting array of cluster values.
k_predictions_2
```

```python
# Create a copy of the DataFrame with the PCA data
coin_pca_df_2 = coin_pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
coin_pca_df_2['predictions'] = k_predictions_2


# Display sample data
coin_pca_df_2.head(5)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
pca_scatter = coin_pca_df_2.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="predictions",
    value_label = "predictions",
    hover_cols = 'coin_id'
).opts(yformatter="%.0f")

pca_scatter
```
